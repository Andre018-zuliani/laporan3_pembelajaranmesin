```python
from sklearn.preprocessing import LabelEncoder
for col in ['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'Species']:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    data
    print (data['Id'],data['SepalLengthCm'],data['SepalWidthCm'],data['PetalLengthCm'],data['Species'])
#AndreZuliani_202131031
```

```python
x = data.drop('Species', axis = 1)
y = data['Species']
#AndreZuliani_202131031
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state =0)
#AndreZuliani_202131031
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
#AndreZuliani_202131031
```

```python
knn.fit(x_train, y_train)
#AndreZuliani_202131031
```

```python
print("Akurasi score sebesar: {}".format(knn.score(x_test,y_test)*100))
#AndreZuliani_202131031
```
